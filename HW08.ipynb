# Question 1
### (a) **Understanding Classification Decision Trees**
A **Classification Decision Tree** is a supervised learning algorithm used for **categorical target variables**. It addresses problems where the goal is to classify data into predefined groups based on a set of input features. Essentially, it predicts which "class" an observation belongs to by splitting the data into subsets based on the values of input features.

#### Real-World Applications of Classification Decision Trees:
1. **Medical Diagnosis**: Determining if a patient has a specific disease based on symptoms and test results (e.g., flu, diabetes).
2. **Customer Segmentation**: Classifying customers into different purchasing behavior groups for targeted marketing.
3. **Fraud Detection**: Identifying fraudulent transactions based on transaction patterns.
4. **Spam Detection**: Classifying emails as spam or not spam based on textual and sender attributes.
5. **Loan Approval**: Predicting whether a loan applicant is likely to repay or default based on financial history.

---

### (b) **Difference in Prediction Mechanisms**

#### **How Classification Decision Trees Make Predictions:**
1. **Structure**: The tree consists of nodes (decision points) and leaves (outcomes or classifications). 
2. **Splitting Criteria**: At each node, the tree uses a metric (like Gini Impurity, Entropy, or Information Gain) to decide which feature and value to split on to maximize separation between classes.
3. **Path to Prediction**: Data is passed through the tree, starting from the root node. At each node, a decision is made based on the input features, routing the observation to a subsequent branch until it reaches a leaf node.
4. **Final Output**: The leaf node assigns the class label. If the tree is probabilistic, it may provide the likelihood of belonging to each class.

#### **How Multiple Linear Regression Makes Predictions:**
1. **Nature of Problem**: Regression predicts a **continuous outcome** (e.g., house prices, stock values).
2. **Model Structure**: It creates a linear equation in the form \( Y = \beta_0 + \beta_1X_1 + \beta_2X_2 + \ldots + \beta_nX_n + \epsilon \), where \( Y \) is the predicted value, \( X_i \) are features, \( \beta_i \) are coefficients, and \( \epsilon \) is error.
3. **Path to Prediction**: It computes the outcome by substituting the feature values into the equation.
4. **Final Output**: The prediction is a single, continuous value.

---

### Key Differences:
| Feature                    | Classification Decision Tree                         | Multiple Linear Regression                         |
|----------------------------|-----------------------------------------------------|--------------------------------------------------|
| **Output Type**            | Class label (categorical)                           | Continuous value (numerical)                     |
| **Prediction Mechanism**   | Sequential splits based on feature values           | Linear combination of features and coefficients  |
| **Evaluation Metrics**     | Accuracy, Precision, Recall, F1 Score, etc.         | R-squared, Mean Squared Error, etc.              |
| **Use Cases**              | Categorization tasks (e.g., fraud detection)        | Prediction of numerical outcomes (e.g., prices)  |

### Expanded Analysis of Metrics with Memorable Examples

To better understand the distinctions between **Accuracy**, **Sensitivity**, **Specificity**, and **Precision**, let’s explore how decision-making supported by each metric varies in different contexts. This exploration will include **distinct examples** and delve deeper into how these metrics align with real-world priorities.

---

### 1. **Accuracy**
**Definition**: Measures the proportion of correctly classified instances (both true positives and true negatives) in the total population.

#### Real-World Application: **General Quality Assessment**
- **Example**: Self-driving car performance.
  - **Scenario**: A self-driving car uses an algorithm to detect pedestrians and obstacles. Both false positives (mistaking a shadow for a person) and false negatives (failing to detect an actual pedestrian) have comparable consequences (e.g., stopping unnecessarily or causing harm).
  - **Why Accuracy?**: Here, the importance is on the **overall performance** of the algorithm, balancing true detections and false errors across all cases.

---

### 2. **Sensitivity (Recall)**
**Definition**: Proportion of actual positives that are correctly identified.

#### Real-World Application: **Life-Critical Medical Screening**
- **Example**: Screening newborns for genetic disorders.
  - **Scenario**: A hospital tests newborns for a rare but treatable condition, such as phenylketonuria (PKU). Missing a positive case (false negative) could mean the baby doesn’t get early treatment, leading to severe outcomes.
  - **Why Sensitivity?**: In this case, **catching every positive case matters more than false alarms**, as further diagnostic tests can filter out false positives. Prioritizing sensitivity minimizes the chance of missing an affected baby.

#### Real-World Application: **Natural Disaster Prediction**
- **Example**: Hurricane early-warning systems.
  - **Scenario**: Meteorological algorithms predict if a storm will escalate into a hurricane. Missing a hurricane’s development is far more dangerous than a false warning.
  - **Why Sensitivity?**: High sensitivity ensures all potential hurricanes are flagged for preparation, even if some predictions turn out to be false positives.

---

### 3. **Specificity**
**Definition**: Proportion of actual negatives that are correctly identified.

#### Real-World Application: **Spam Filtering**
- **Example**: Email service classifying emails as spam.
  - **Scenario**: A spam filter must correctly allow legitimate emails (true negatives) through while blocking spam (false positives). Incorrectly blocking critical emails (false positives) can cause users to lose important information.
  - **Why Specificity?**: A high specificity ensures fewer legitimate emails are blocked as spam, even if a few spam messages occasionally slip through.

#### Real-World Application: **Criminal Background Checks**
- **Example**: Screening job applicants for a criminal record.
  - **Scenario**: A company screens job applicants for past criminal records. Wrongly marking someone as having a criminal record (false positive) could harm their career prospects unfairly.
  - **Why Specificity?**: High specificity minimizes false positives, ensuring innocent individuals aren’t flagged incorrectly.

---

### 4. **Precision**
**Definition**: Proportion of positive predictions that are correct.

#### Real-World Application: **Fraud Detection in Banking**
- **Example**: Identifying fraudulent transactions.
  - **Scenario**: A bank uses algorithms to flag suspicious transactions for manual review. Investigating false positives wastes resources and slows down legitimate transactions.
  - **Why Precision?**: High precision ensures that flagged transactions are highly likely to be fraudulent, reducing unnecessary investigations and customer inconvenience.

#### Real-World Application: **Targeted Marketing**
- **Example**: Personalized product recommendations.
  - **Scenario**: An e-commerce platform recommends products based on user preferences. Showing irrelevant recommendations (false positives) reduces user satisfaction.
  - **Why Precision?**: Precision ensures the algorithm only shows products that are likely to interest the customer, improving engagement and conversion rates.

---

### Comparison Table of Metrics and Decision-Making Contexts

| **Metric**       | **Priority**                                                                 | **Decision-Making Context**                                         | **Example**                                      |
|-------------------|-----------------------------------------------------------------------------|---------------------------------------------------------------------|-------------------------------------------------|
| **Accuracy**      | Overall performance, balanced priorities                                    | Equal cost for false positives and negatives                        | Self-driving car pedestrian detection           |
| **Sensitivity**   | Avoiding false negatives; critical to catching every positive case         | High cost of missing positives                                      | Newborn genetic disorder screening              |
| **Specificity**   | Avoiding false positives; critical to protecting true negatives            | High cost of incorrectly identifying negatives                      | Spam filtering                                  |
| **Precision**     | Ensuring positive predictions are correct                                  | High cost of acting on false positives                              | Fraud detection in banking                      |

---

### How the Metrics Complement Each Other
- **Accuracy** provides a **broad performance measure** but may fail in imbalanced datasets.
- **Sensitivity** is ideal when **false negatives** are unacceptable (e.g., critical health diagnoses).
- **Specificity** shines when **false positives** need to be minimized (e.g., criminal screenings).
- **Precision** prioritizes the correctness of **positive predictions**, crucial for resource-intensive follow-ups.

### Explanation of Predictions for `clf2`
1. **Decision Splits**:
   - The tree uses the features `NumPages`, `Thick`, and `List Price` to create splits at various thresholds.
   - At each node, the condition (e.g., `NumPages <= 200`) determines whether a sample follows the left or right branch.

````python
from sklearn.metrics import confusion_matrix, accuracy_score, recall_score, ConfusionMatrixDisplay

# Prepare test data
y_test = pd.get_dummies(ab_reduced_noNaN_test["Hard_or_Paper"])['H']
X_test_clf = ab_reduced_noNaN_test[['List Price']]
X_test_clf2 = ab_reduced_noNaN_test[['NumPages', 'Thick', 'List Price']]

# Generate predictions
y_pred_clf = clf.predict(X_test_clf)
y_pred_clf2 = clf2.predict(X_test_clf2)

# Confusion matrix and metrics for clf
conf_matrix_clf = confusion_matrix(y_test, y_pred_clf)
accuracy_clf = accuracy_score(y_test, y_pred_clf)
sensitivity_clf = recall_score(y_test, y_pred_clf)
specificity_clf = conf_matrix_clf[0, 0] / (conf_matrix_clf[0, 0] + conf_matrix_clf[0, 1])

# Display confusion matrix for clf
ConfusionMatrixDisplay(conf_matrix_clf, display_labels=["Paperback", "Hardcover"]).plot()
plt.title("Confusion Matrix for clf")
plt.show()

# Confusion matrix and metrics for clf2
conf_matrix_clf2 = confusion_matrix(y_test, y_pred_clf2)
accuracy_clf2 = accuracy_score(y_test, y_pred_clf2)
sensitivity_clf2 = recall_score(y_test, y_pred_clf2)
specificity_clf2 = conf_matrix_clf2[0, 0] / (conf_matrix_clf2[0, 0] + conf_matrix_clf2[0, 1])

# Display confusion matrix for clf2
ConfusionMatrixDisplay(conf_matrix_clf2, display_labels=["Paperback", "Hardcover"]).plot()
plt.title("Confusion Matrix for clf2")
plt.show()

# Report metrics
results = pd.DataFrame({
    "Model": ["clf", "clf2"],
    "Accuracy": [accuracy_clf, accuracy_clf2],
    "Sensitivity": [sensitivity_clf, sensitivity_clf2],
    "Specificity": [specificity_clf, specificity_clf2],
})

print(results)
````

# Question 9
In **linear regression**, coefficients represent the direct effect of each predictor variable on the outcome. Specifically, a coefficient indicates how much the dependent variable is expected to change for a one-unit increase in the predictor variable, assuming all other variables are held constant. This allows for a clear, additive interpretation of the relationships between predictors and the outcome.

In **decision trees**, feature importances reflect the relative contribution of each feature to the model's splits and overall predictive performance. Unlike regression coefficients, these values do not provide a direct or linear effect but instead quantify how much each feature reduces impurity (e.g., Gini index or entropy) across all splits in the tree. The interpretation is more global, showing which features were most influential overall, but not their precise impact on individual predictions.
